##### import the titanic dataset

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dftitanic=pd.read_csv('titanic.csv')
```

```python
dftitanic.head()
```

##### 1. Write a Python program to display a bar chart of the fare of class.

```python
sns.barplot(x='class', y='fare', data=dftitanic)
plt.title('Titanic fare according to class')
plt.show()
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python

sns.barplot(x='fare', y='class', data=dftitanic,ci=None,orient = "h")
plt.title('Titanic fare according to class')
plt.show()
```

###### 3. Write a Python program to display a bar chart of the fare of class. Use different color for each bar.

```python
plt.bar(dftitanic[ 'class'],dftitanic['fare'], color=['red', 'black', 'green'])
plt.title('fare of class')
plt.show
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
dftitanic["class"].value_counts().plot.pie(autopct="%.1f%%")
plt.figure(figsize=(10,5))
plt.show()
```
